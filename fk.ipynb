```python
import pandas as pd
import numpy as np
```

```python
fake=pd.read_csv('R:\\Personal\\College\\Ai - COE\\Project\\fake.csv')
true=pd.read_csv('R:\\Personal\\College\\Ai - COE\\Project\\True.csv')
```

```python
fake['label']=0
true['label']=1
```

```python
news=pd.concat([fake,true],axis=0)
news.head()
```

```python
news.isnull().sum()
```

```python
news=news.drop(['title','subject','date'],axis=1)
```

```python
news.head()
```

```python
news=news.sample(frac=1)        #reshuffling
```

```python
news.head()
```

```python
news.reset_index(inplace=True)
```

```python
news.head()
```

```python
news.drop(['index'],axis=1,inplace=True)
```

```python
news.head()
```

```python
import re
```

```python
def wordopt(text):      #deals url information to words
    text=text.lower()                                   #convert to lowercase
    text=re.sub(r'https?://\S+|www\.\S+','',text)       #remove url
    text=re.sub(r'<.*?>0','',text)                      #remove html tags
    text=re.sub(r'[^\w\s]','',text)                     #remove punctuation
    text=re.sub(r'\d','',text)                          #remove digits
    text=re.sub(r'\n',' ',text)                         #remove newline characters
    return text
```

```python
news['text']=news['text'].apply(wordopt)
news['text']
```

```python
x=news['text']
y=news['label']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization=TfidfVectorizer()
```

```python
xv_train=vectorization.fit_transform(x_train)
```

```python
xv_test=vectorization.transform(x_test)
```

```python
xv_test
```

```python
xv_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR=LogisticRegression()
```

```python
LR.fit(xv_train,y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
from sklearn.metrics import classification_report 
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC=DecisionTreeClassifier()
```

```python
DTC.fit(xv_train,y_train)
```

```python
pred_dtc =DTC.predict(xv_test)
```

```python
DTC.score(xv_test,y_test)
```

```python
print(classification_report(y_test, pred_dtc))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rtc=RandomForestClassifier()
```

```python
rtc.fit(xv_train,y_train)
```

```python
predict_rtc=rtc.predict(xv_test)
```

```python
rtc.score(xv_test,y_test)
```

```python
print(classification_report(y_test,predict_rtc))
```

```python
gbc=GradientBoostingClassifier()
```

```python
pred_gbc=gbc.predict(xv_test)
```

```python
def output_label(n):
    if n==0:
        return "It is a Fake News"
    elif n==1:
        return "It is a Genuine News"
```

```python
def manual_testing(news):
    testing_news={"text":[news]}        #corrected syntax for defining dictionary
    new_def_test=pd.DataFrame(testing_news)
    new_def_test["text"]=new_def_test["text"].apply(wordopt)
    new_x_test=new_def_test['text']
    new_xv_test=vectorization.transform(new_x_test) #assuming 'vectorization' is your vectorizer object
    pred_lr=LR.predict(new_xv_test)
    pred_gbc=gbc.predict(new_xv_test)
    predict_rtc=rtc.predict(new_xv_test)
    return "\n\nLR Prediction:{}  \nGBC Prediction:{}  \nRFC Prediction: {}".format(
        output_label(pred_lr[0]), output_label(pred_gbc[0]), output_label(predict_rtc[0])
    )
```

```python
news_article=str(input())
```

```python
manual_testing(news_article)
```
